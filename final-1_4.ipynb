```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Perceptron:
#Picks x(t) deterministically(the first misclassfiied)    
    def __init__(self, x, y, w):
        self.x = x
        self.y = y
        self.w = w
        
    def predict(self):
        h = np.dot(self.x,self.w.T)
        h_sign = np.zeros(h.size)
        for i in range(h_sign.size):
            if h[i] > 0:
                h_sign[i] = 1
            else:
                h_sign[i] = -1
        return h_sign
    
    def misclassified(self, h):
        miscls = []
        for i in range(y.size):
            if y[i] != h[i]:
                miscls.append(i)
        miscls = np.array(miscls)
        return miscls
        
    def train(self):
        count = 0
        while True:
            h = self.predict()
            miscls = self.misclassified(h)
            if miscls.size > 0:
                #pick the first misclassified
                t = miscls[0]
                #t = np.random.choice(miscls)
                self.weight_update(self.y[t],self.x[t])
            else:
                return self.w,count
            count = count + 1
            
    def weight_update(self,y_t,x_t):
        self.w = self.w + y_t*x_t
```

```python
def sign(x,w):
    return np.sign(np.dot(w,x.T))
#generate data set where x_n in R^2    
def generate_data(size):
    n=size
    w = np.random.randint(low=-100,high= 100,size = 3)
    y = np.zeros(n)
    x = []
    for i in range(n):
        x_i = [1,np.random.randint(0, 100), np.random.randint(0, 100)]
        x.append(x_i)
    x = np.array(x)
    y = sign(x,w)
    return x,y,w

def abline(slope, intercept, col,label):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals,col,label = label)
```

```python
#part a)
x, y, ideal_weight = generate_data(20)
plt.figure(figsize=(15,15))
#Class 1
plt.plot(x[np.where(y==1),1],x[np.where(y==1),2], 'ro')
#Class 2
plt.plot(x[np.where(y==-1),1],x[np.where(y==-1),2], 'bo')
#Target function in yellow
abline(slope = -(ideal_weight[1]/ideal_weight[2]),intercept = -(ideal_weight[0]/ideal_weight[2]),col='-y',label="target function")
plt.axis([0, 110, 0, 110])
plt.xlabel('x1', fontsize=25)
plt.ylabel('x2', fontsize=25)
plt.legend(loc='best', fontsize=20)
plt.title('a) Linearly separable data set of size 20', fontsize=25)
plt.show()
```

```python
#Part b)
w = np.zeros(3)
newP = Perceptron(x, y, w)
weights,count = newP.train()
print("Number of updates that the algorithm takes before converging:"+  str(count) )
plt.figure(figsize=(15,15))
#Class 1
plt.plot(x[np.where(y==1),1],x[np.where(y==1),2], 'ro')
#Class 2
plt.plot(x[np.where(y==-1),1],x[np.where(y==-1),2], 'bo')
#Hypothesis in green
abline(slope = -(weights[1]/weights[2]),intercept = -(weights[0]/weights[2]),col='-g',label="hypothesis")
#Target function in yellow
abline(slope = -(ideal_weight[1]/ideal_weight[2]),intercept = -(ideal_weight[0]/ideal_weight[2]),col='-y',label="target function")
plt.axis([0, 110, 0, 110])
plt.xlabel('x1', fontsize=25)
plt.ylabel('x2', fontsize=25)
plt.legend(loc='best', fontsize=20)
plt.title('b) Linearly separable data set of size 20', fontsize=25)
plt.show()
```

```python
#Part f)
class Perceptron2:
#Picks x(t) randomly instead of deterministically    
    def __init__(self, x, y, w):
        self.x = x
        self.y = y
        self.w = w
        
    def predict(self):
        h = np.dot(self.x,self.w.T)
        h_sign = np.zeros(h.size)
        for i in range(h_sign.size):
            if h[i] > 0:
                h_sign[i] = 1
            else:
                h_sign[i] = -1
        return h_sign
    
    def misclassified(self, h):
        miscls = []
        for i in range(y.size):
            if y[i] != h[i]:
                miscls.append(i)
        miscls = np.array(miscls)
        return miscls
        
    def train(self):
        count = 0
        while True:
            h = self.predict()
            miscls = self.misclassified(h)
            if miscls.size > 0:
                #pick a random misclassified
                t = np.random.choice(miscls)
                self.weight_update(self.y[t],self.x[t])
            else:
                return self.w,count
            count = count + 1
            
    def weight_update(self,y_t,x_t):
        self.w = self.w + y_t*x_t
#generate data set where x_n in R^10        
def generate_data10(size):
    n=size
    w = np.random.randint(low=-100,high= 100,size = 11)
    y = np.zeros(n)
    x = []
    for i in range(n):
        x_i = [1,np.random.randint(0, 100),np.random.randint(0, 100),
               np.random.randint(0, 100),np.random.randint(0, 100),
               np.random.randint(0, 100),np.random.randint(0, 100),
               np.random.randint(0, 100),np.random.randint(0, 100),
               np.random.randint(0, 100),np.random.randint(0, 100)]
        x.append(x_i)
    x = np.array(x)
    y = sign(x,w)
    return x,y,w
```

```python
#Part f
w = np.zeros(11)
x, y, ideal_weight = generate_data10(1000)
newP = Perceptron2(x, y, w)
weights,count = newP.train()
print("Number of updates that the algorithm takes before converging:"+  str(count) )
```
